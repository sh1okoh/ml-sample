```python
in_file = 'data.csv'

with open(in_file, "r", encoding="Shift_JIS") as fr:
  lines = fr.readlines()

converted_lines = ["年月日,気圧\n"] + lines[5:]
result = "".join(map(lambda value: value.replace(',8,1', ''), converted_lines)).strip()
print(result)
```
